```python
df = df.drop_duplicates()
```

```python
print(df['Gender'].unique())
```

```python
def clean_gender(val):
    if str(val).strip().lower() in ['male', 'm', '1']:
        return 'Male'
    elif str(val).strip().lower() in ['female', 'f', '0']:
        return 'Female'
    else:
        return np.nan
```

```python
df['Gender_Clean'] = df['Gender'].apply(clean_gender)
```

```python
df['Gender_Clean']
```

```python
print(df['Gender_Clean'].unique())
```

```python
df['Salary_Clean'] = df['Salary'].astype(str)
df['Salary_Clean'] = df['Salary_Clean'].str.replace('rupees','', regex=False)
df['Salary_Clean'] = df['Salary_Clean'].str.replace('k','000', regex=False)
df['Salary_Clean'] = df['Salary_Clean'].str.replace(',','', regex=False)
df['Salary_Clean'] = pd.to_numeric(df['Salary_Clean'], errors='coerce')
```

```python
df['Salary_Clean']
```

```python
df['Age_Clean'] = df['Age'].astype(str).str.replace(' years','')
df['Age_Clean'] = pd.to_numeric(df['Age_Clean'], errors='coerce')
```

```python
df.loc[(df['Age_Clean'] < 18) | (df['Age_Clean'] > 80), 'Age_Clean'] = np.nan
```

```python
df['Age_Clean']
```

```python
df['Experience_Clean'] = df['Experience_Years'].astype(str)
df['Experience_Clean'] = df['Experience_Clean'].str.replace(' yrs','')
df['Experience_Clean'] = df['Experience_Clean'].replace('fresher', '0')
df['Experience_Clean'] = pd.to_numeric(df['Experience_Clean'], errors='coerce')
```

```python
df['Experience_Clean']
```

```python
print("Null before:", df['Age_Clean'].isnull().sum())

df['Age_Clean'].fillna(df['Age_Clean'].mean(), inplace=True)

print("Null after:", df['Age_Clean'].isnull().sum())
```

```python
df['Salary_Clean'].fillna(df['Salary_Clean'].median(), inplace=True)
df['Experience_Clean'].fillna(df['Experience_Clean'].median(), inplace=True)
```

```python
df['Age_Clean']
```

```python
df['Salary_Clean']
```

```python
df['Experience_Clean']
```

```python
cat_cols = ['Gender_Clean','City','Department','Education','Performance_Rating']

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
print(df.columns)
```

```python
df['Score_Clean'] = df['Performance_Score'].astype(str).str.replace('%','')
df['Score_Clean'] = pd.to_numeric(df['Score_Clean'], errors='coerce')
```

```python
df.loc[(df['Score_Clean'] < 0) | (df['Score_Clean'] > 100), 'Score_Clean'] = np.nan
```

```python
print(df[['Performance_Score','Score_Clean']].head())
```

```python
df['Loan_Approved'] = df['Loan_Approved'].fillna(method='ffill')
df['Loan_Approved'] = df['Loan_Approved'].fillna(method='bfill')
```

```python
print(df['Loan_Approved'].isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender_Encoded'] = le.fit_transform(df['Gender_Clean'])

print(dict(zip(le.classes_, le.transform(le.classes_))))
print(df['Gender_Encoded'].value_counts())
```

```python
df['Loan_Encoded'] = df['Loan_Approved'].map({'Yes':1, 'No':0})
print(df[['Loan_Approved','Loan_Encoded']].head(10))
```

```python
dept_dummies = pd.get_dummies(df['Department'], prefix='Dept')

print("Before shape:", df.shape)

df = pd.concat([df, dept_dummies], axis=1)
df.drop('Department', axis=1, inplace=True)

print("After shape:", df.shape)
print(dept_dummies.columns)
```

```python
df['City'] = df['City'].str.strip().str.title()

city_dummies = pd.get_dummies(df['City'], prefix='City', drop_first=True)
df = pd.concat([df, city_dummies], axis=1)
df.drop('City', axis=1, inplace=True)

print("New city columns:", len(city_dummies.columns))
```

```python
from sklearn.preprocessing import OrdinalEncoder

edu_order = ['High School','Diploma',"Bachelor's","Master's",'PhD']

oe = OrdinalEncoder(categories=[edu_order])
df['Education_Encoded'] = oe.fit_transform(df[['Education']]) + 1

print(pd.crosstab(df['Education'], df['Education_Encoded']))
```

```python
rating_map = {
    'Poor':1,
    'Average':2,
    'Good':3,
    'Excellent':4
}

df['Rating_Encoded'] = df['Performance_Rating'].map(rating_map)

print(pd.crosstab(df['Performance_Rating'], df['Rating_Encoded']))
```

```python
Q1 = df['Salary_Clean'].quantile(0.25)
Q3 = df['Salary_Clean'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Outliers below:", (df['Salary_Clean'] < lower).sum())
print("Outliers above:", (df['Salary_Clean'] > upper).sum())

# Capping
df['Salary_Capped'] = np.where(df['Salary_Clean'] < lower, lower,
                       np.where(df['Salary_Clean'] > upper, upper,
                                df['Salary_Clean']))

print(df['Salary_Capped'].min(), df['Salary_Capped'].max())
```

```python
# Join date
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')
df['Join_Date'].fillna(df['Join_Date'].median(), inplace=True)

today = pd.Timestamp.today()

df['Tenure_Years'] = ((today - df['Join_Date']).dt.days / 365).round(1)
df['Join_Year'] = df['Join_Date'].dt.year
df['Join_Quarter'] = df['Join_Date'].dt.quarter
```

```python
df['Tenure_Years']
```

```python
df['Salary_Per_Year_Exp'] = (df['Salary_Capped'] / (df['Experience_Clean'] + 1)).round(2)

df['Score_Salary_Ratio'] = (df['Score_Clean'] / df['Salary_Capped'] * 1000).round(4)

df['Is_Senior'] = np.where(df['Age_Clean'] >= 40, 1, 0)

df['Exp_Age_Ratio'] = (df['Experience_Clean'] / df['Age_Clean']).round(3)

print(df[['Salary_Per_Year_Exp','Score_Salary_Ratio','Is_Senior','Exp_Age_Ratio']].describe())
```

```python
bins = [18,30,42,54,80]
labels = ['Young','Mid-Career','Senior','Veteran']

df['Age_Group'] = pd.cut(df['Age_Clean'], bins=bins, labels=labels)

age_map = {'Young':1,'Mid-Career':2,'Senior':3,'Veteran':4}
df['Age_Group_Encoded'] = df['Age_Group'].map(age_map)

print(df['Age_Group'].value_counts())
```

```python
df['Salary_Bracket'] = pd.qcut(df['Salary_Capped'], 4,
                               labels=['Low','Lower-Mid','Upper-Mid','High'])

print(df['Salary_Bracket'].value_counts())
print(pd.crosstab(df['Salary_Bracket'], df['Loan_Approved']))
```

```python
bins = [0,50,60,70,80,90,100]
labels = ['F','D','C','B','A','S']

df['Score_Grade'] = pd.cut(df['Score_Clean'], bins=bins, labels=labels)

grade_dummies = pd.get_dummies(df['Score_Grade'], prefix='Grade')
df = pd.concat([df, grade_dummies], axis=1)

print(df['Score_Grade'].value_counts())
```

```python
final_cols = [
    'Age_Clean','Salary_Capped','Score_Clean','Experience_Clean','Tenure_Years',
    'Salary_Per_Year_Exp','Is_Senior','Exp_Age_Ratio','Age_Group_Encoded',
    'Rating_Encoded','Education_Encoded','Gender_Encoded','Loan_Encoded'
]

dept_cols = [col for col in df.columns if col.startswith('Dept_')]

df_final = df[final_cols + dept_cols]

print(df_final.isnull().sum())
print(df_final.shape)
print(df_final.head())
```
